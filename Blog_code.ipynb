```python
# 3.What is the relationship between the most widely used language in United States and the Salary for that langanguage?
```

```python
"""
 3.Prepare Data
The data will be prepared here by selecting the desired columns and only using the rows where a full-stack developer was taking the survey
All rows were dropped where the DevType column was  not equal to "Full-stack developer" as I was only interested in Full-stack developer.
If the row had ConvertedSalary or LanguageWorkedWith as Nan then the row would be dropped. 
This is becase I was trying to find a correlation between the two and therefore they needed to exist. 
This was the only way in which I needed to handle Nan values and categorical values.
"""
```

```python
#Find the relationship between the most common langauge and the salary given for that language
Lang_Hsalary = US_2018[['DevType', 'LanguageWorkedWith', 'ConvertedSalary']]  #Make  a new table from US_2018 with DevtYoe,LAnguage and Salary as the columns

Lang_Hsalary = Lang_Hsalary[Lang_Hsalary.DevType == "Full-stack developer"]   #Filter the data and keep only those data where the devtype of a programmer is full stack

Lang_Hsalary= Lang_Hsalary.dropna(subset=['ConvertedSalary', 'LanguageWorkedWith'], axis=0)  # Drop the rows where salary and devtype is nan
Lang_Hsalary.head()
```

```python
""" 4.Data Modelling """
```

```python
def split_column(column_split, delimeter, column_store, df):
    """Method to split a columns values by a provided delimeter while
    retaining the associated value of another column.
    
    :param str column_split: column whose values are to be split
    :param str delimeter: delimeter to split by
    :param str columns_store: column where the  data must be stored along split rows
    :param DataFrame df: dataframe to use
    """
    
    split_col = pd.DataFrame(columns = [column_split, column_store])
    for index, row in df.iterrows():
        columns = row[column_split].split(delimeter)
        for c in columns:
            split_col.loc[len(split_col)] = [c, row[column_store]]
    return split_col
```

```python
split_languages = split_column("LanguageWorkedWith", ";", "ConvertedSalary", Lang_Hsalary)
split_languages.head()
```

```python
sal_table = split_languages.groupby(['LanguageWorkedWith']).median()['ConvertedSalary'].sort_values(ascending=False)
plt.figure(figsize=(10,10))
sal_table.plot(kind = 'bar')
plt.xlabel("LanguageWorkedWith")
plt.ylabel("Annual Salary in USD")
```

```python
""" 
5.Evaluate the result:
From the above graph we can make a unique conclusion that the language which is least popular i.e. which is used the least
is the language which is paid the most!!Where as javascript being the most popular language is paid less than Hack(least popular)"""
```

```python
# 4. What is the female ratio of programmers in the United States.(w.r.t employement)?
```

```python
"""
3. Prepare Data
The same data used for question 1 is used here.
The data will be prepared here by selecting the desired columns and only using the rows where country is UNITED STATES.
All rows were dropped where the Country column did not equal "United States".
This was the only way in which I needed to handle Nan values and categorical values."""
```

```python
""" 4. Data Modelling """
```

```python
#To select the column wherein we have to find the female/male ratio
female_ratio = US_2018.pivot_table(index='Employment', columns='Gender', aggfunc='size', fill_value=0) 
sums = female_ratio[['Female', 'Male']].sum(axis=1)                                #the total proportion of male and female
female_ratio['FemaleRatio'] = round(100 * female_ratio['Female'] / sums , 1)     # the female_rati
female_ratio['FemaleRatio']
```

```python
"""Create a new dataframe with Employment and FemaleRatio as columns 
from the data recieved in female_ratio."""

data={'Employment':['Employed full-time','Employed part-time ','Independent contractor, freelancer, or self-employed ','Not employed, and not looking for work ','Not employed, but looking for work ','Retired '],'FemaleRatio':['8.4','10.9','8.1','5.9','10.3','11.1']}
US_Genratio= pd.DataFrame(data)
US_Genratio
```

```python
 #plot employment v/s female ratio graph using sns plot
plt.figure(figsize=(6,6))
sns.barplot(x = 'FemaleRatio', y = 'Employment',data = US_Genratio)
plt.title(' Ratio of Female Programmers w.r.t employment in United States', size = 16)
plt.xlabel("Female Ratio", fontsize = 14)
plt.ylabel("Employment Status", fontsize = 14)

plt.show()
```
